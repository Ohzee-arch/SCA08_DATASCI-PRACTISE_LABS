```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv("nba.csv")
```

```python
data.head()
```

```python
data.corr(method="pearson")
```

```python
positions = list(data["Position"].unique())
```

```python
for i, position in enumerate(positions):
    data.replace(position, i, inplace=True)
```

```python
data.head()
```

```python
data["Number"] =  data["Number"].astype(int)
```

```python
data["Number"].dtype
```

```python
data["Position"] =  data["Position"].astype(int)
```

```python
data["Position"].dtype
```

```python
data["Weight"] =  data["Weight"].astype(int)

data["Weight"].dtype
```

```python
data["Salary"] = data["Salary"].astype(int)

data["Salary"].dtype
```

```python
data.corr("pearson")
```

```python
#changing the column "Height" to numerical data for better analysis

heights = list(data["Height"].unique())
```

```python
for i, height in enumerate(heights):
    data.replace(height, i, inplace=True)
```

```python
data.head()
```

```python
data.corr("pearson")
```

```python
#changing the columns "Team" and "College" to numerical data for better analysis


teams = list(data["Team"].unique())

colleges = list(data["College"].unique())
```

```python
for i, team in enumerate(teams):
    data.replace(team, i, inplace=True)
    
for i, college in enumerate(colleges):
    data.replace(college, i, inplace=True)
```

```python
data.head()
```

```python
data.corr("pearson")
```

```python
#From the overall correlation analysis displayed above, we understand that Team has a little correlation with College, Number has a little correlation with Position, Position has more correlation with weight, Age has just a little correlation Salary. 
```

```python
result = pd.merge(left, right, on=['A','B'], how='outer')
```
